```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Pinky", "Rinky", np.nan, "Latha", "Ruhi"],
    "Age": [25, np.nan, 30, 35, np.nan],
    "Salary": [50000, 60000, np.nan, 70000, 80000],
    "Department": ["HR", "IT", "Finance", np.nan, "IT"]
}

df = pd.DataFrame(data)
print(df)
```

```python
d1 = df.copy()
print(d1)
```

```python
from sklearn.impute import SimpleImputer
```

```python
num_imputers = SimpleImputer(strategy = 'mean')
num_cols = d1.select_dtypes( include = [np.number]).columns
d1[num_cols] = num_imputers.fit_transform(d1[num_cols])
d1.head()
```

```python
num_cat = SimpleImputer(strategy = 'most_frequent')
cat_cols= d1.select_dtypes( include = ['object']).columns
d1[cat_cols] = num_cat.fit_transform(d1[cat_cols])
d1.head()
```

```python
from sklearn.impute import KNNImputer
```
