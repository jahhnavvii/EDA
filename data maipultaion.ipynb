```python
# Comprehensive analysis of your data
def analyze_financial_data(df):
    # Filter financial performance data
    financial_df = df[df['Variable_category'] == 'Financial performance']
    
    # Summary by industry
    industry_summary = financial_df.groupby('Industry_name_NZSIOC').agg({
        'Value': ['sum', 'mean', 'std', 'count']
    }).round(2)
    
    # Top variables by value
    top_variables = financial_df.groupby('Variable_name')['Value'].sum().nlargest(10)
    
    # Industry ranking
    industry_ranking = financial_df.groupby('Industry_name_NZSIOC')['Value'].sum().sort_values(ascending=False)
    
    return {
        'industry_summary': industry_summary,
        'top_variables': top_variables,
        'industry_ranking': industry_ranking
    }

# Run analysis
results = analyze_financial_data(df)

print("Industry Summary:")
print(results['industry_summary'])
print("\nTop Variables:")
print(results['top_variables'])
print("\nIndustry Ranking:")
print(results['industry_ranking'])
```
